### Planck's Law

$$ B_\nu(\nu, T) = \frac{2 h \nu^3}{c^2} \frac{1}{\text{exp}(\frac{h\nu}{k_BT}) - 1}$$

B<sub>$\nu$</sub>: spectral brightness - power emitted per unit area per unit frequency per steradian of solid angle by a blackbody

h: Planck's constant - 6.63E-34 J s

$\nu$: frequency (Hz; s<sup>-1</sup>)

k<sub>B</sub>: Boltzmann's constant - 1.38E-23 J K<sup>-1</sup>

c: speed of light in a vacuum - 3.00E8 m s<sup>-1</sup>

T: temperature (Kelvin - K)

Rayleigh Jean's Limit of Planck's equation. Where

$$\frac{h\nu}{kT} << 1$$

$$B_\nu(\nu,
T) = \frac{2k_BT\nu^2}{c^2}$$

### Specific Intensity Formula

$$I_\nu = \frac{dP}{(cos\theta d\sigma) d\nu d\Omega}$$

I<sub>$\nu$</sub>: specific source intensity - W m<sup>-2</sup> Hz<sup>-1</sup> sr<sup>-1</sup>

d<sub>$\sigma$</sub>: an infinitesimal surface area (e.g. of the detector/antenna - m<sup>2</sup>

$\theta$: the angle between a "ray" of radiation and the normal to the surface (measured from the vertical axis) - radians

d$\Omega$: an infinitesimal solid angle measured from the observer's location - sr

$\nu$: frequency - Hz

### Flux Density

$$S_\nu = \int_{\Omega} I_\nu(\theta, \phi) cos\theta d\Omega$$

S<sub>$\nu$</sub>: flux density - W m<sup>-2</sup> Hz<sup>-1</sup>

Integrated over the entire source - in our case: 0 < $\phi$ < $\pi$/2 and 0 < $\theta$ < 2$\pi$

### Jansky Conversion

1 jansky = 1 Jy = 10<sup>-26</sup> W m<sup>-2</sup> Hz<sup>-1</sup>

### Brightness Temperature

Radio astronomers often find it convenient to specify the spectral brightness I<sub>$\nu$</sub>, even if I<sub>$\nu$</sub> $\neq$ B<sub>$\nu$</sub>, in terms of the equivalent Rayleigh-Jeans brightness temperature T<sub>b</sub>.

$$T_b(\nu) = \frac{I_\nu(\nu) c^2}{2k_B\nu^2}$$

Brightness temperature is just another way to specify power per unit solid angle per unit bandwidth in terms of the Rayleigh-Jeans approximation. It is convenient because radio telescopes are often calibrated by absorbers or "loads" of known temperature and because the temperature of a radio source is frequently a quantity of physical interest. Beware that brightness temperature is NOT the same as physical temperature.

### Noise Generated by a Warm Resistor - Antennas

An antenna is a passive device that converts electromagnetic radiation in electrical currents in wires. The noise generated by a resistor is indistinguishable from the noise coming from a receiving antenna surrounded by blackbody radiation of the same temperature. Warm resistors are useful in radio astronomy as standards for calibrating antennas and receivers, so the noise power per unit bandwidth received by a radio telescope is often described as in terms of the Rayleigh-Jeans ANTENNA TEMPERATURE. The antenna temperature of a receiving antenna is defined as the temperature of an ideal resistor that would generate the same Rayleigh-Jeans noise power per unit bandwidth as appears at the antenna output. Like brightness temperature, antenna temperature is not a physical temperature.

The Nyquist formula determines the total spectral power P<sub>$\nu$</sub> generated by a resistor is

$$P_\nu(\nu, T) = \frac{h\nu}{\text{exp}(\frac{h\nu}{k_B T}) - 1}$$

The Rayleigh-Jeans Limit of this equation where

$$\frac{h\nu}{kT} << 1$$

becomes

$$P_\nu(\nu, T) = k_B T$$

where T is the physical temperature of the resistor

### Effective Area of a Receiving Antenna

$$A_e = \frac{2P_\nu}{S_\nu}$$

A<sub>e</sub>: effective area of an antenna - m<sup>2</sup>

P<sub>$\nu$</sub>: spectral power

S<sub>$\nu$</sub>: total flux density from an unpolarized point source

The total spectral power from all directions collected in one polarization is

$$P_\nu(\nu, T) = \frac{1}{2} \int_\Omega A_e(\nu, \theta, \phi) B_\nu(\nu, T, \theta, \phi) d\Omega$$

$$A_e(\lambda, \theta, \phi) = \frac{\lambda ^2 G_{ant}(\lambda, \theta, \phi)}{4\pi}$$

$$A_e(\nu, \theta, \phi) = \frac{c^2 G_{ant}(\nu, \theta, \phi)}{4\pi\nu^2}$$

G: antenna gain

### Antenna Temperature

A convenient practical unit for the power output per unit frequency from a receiving antenna is the antenna temperature T<sub>A</sub>. Antenna temperature has nothing to do with the physical temperature of the antenna measured by a thermometer; it is only the temperature of a matched resistor whose thermally generated power per unit frequency in the low-frequency Nyquist approximation equals that produced by the antenna.

$$T_A = \frac{P_\nu}{k_B} = \frac{A_e S_\nu}{2k_B}$$

$$T_A = \frac{1}{\lambda^2} \int_{\Omega} A_e(\theta, \phi) T_b(\theta, \phi) d\Omega$$

$$T_A = \frac{\nu^2}{c^2} \int_{\Omega} A_e(\theta, \phi) T_b(\theta, \phi) d\Omega$$

This is the equation we actually want to use - but can't until we get real measurements (not FEKO effective heights)

Although we are going to associate the brightness temperature with a single source (the galactic signal), the instrument "captures" energy from all external bright sources (e.g. the ground, the sun). Furthermore, energy from the galaxy may be diminished due to absorption in the atmosphere (although probably not at the frequencies we're dealing with here).

Recall,

$$T_b(\nu) = \frac{I_\nu(\nu) c^2}{2k_B\nu^2}$$

and

$$A_e(\nu, \theta, \phi) = \frac{c^2 G_{ant}(\nu, \theta, \phi)}{4\pi\nu^2}$$

So,

$$T_A(\nu) = \frac{\nu^2}{c^2} \int_{\Omega} \frac{c^2 G_{ant}(\nu, \theta, \phi)}{4\pi\nu^2} \frac{I_\nu(\nu, \theta, \phi) c^2}{2k_B\nu^2} d\Omega$$

$$T_A(\nu) = \frac{c^2}{8\pi k_B\nu^2} \int_{\Omega} G_{ant}(\nu, \theta, \phi) I_\nu(\nu, \theta, \phi) d\Omega$$

Recall, this is the corresponding temperature measured at the base of our antenna.

### Units

T<sub>A</sub>: antenna temperature - K

c: speed of light - m s<sup>-1</sup>

$\nu$: frequency - s<sup>-1</sup>

k<sub>B</sub>: Boltzmann's constant - J K<sup>-1</sup>

G<sub>ant</sub>: antenna gain - dimensionless

I<sub>$\nu$</sub>: specific source intensity - W m<sup>-2</sup> Hz<sup>-1</sup> sr<sup>-1</sup> = J m<sup>-2</sup> sr<sup>-1</sup>

### Antenna Effective Height and Gain

$$G_{ant}(\nu, \theta, \phi) = \frac{4 \pi \nu^2 \eta h_{eff}^2(\nu, \theta, \phi)} {R_L c^2} $$

The effective heights of the antennas are calculated in a software packaged called FEKO.

### Units

G<sub>ant</sub>: antenna gain - dimensionless

c: speed of light - 3*10<sup>8</sup> m s<sup>-1</sup>

$\nu$: frequency - s<sup>-1</sup>

$\eta$: impedence of the propagating medium (free space) - 377 $\Omega$

R<sub>L</sub>: antenna/load impedence - $\Omega$

h<sub>eff</sub>: antenna effective height - m

So we have...

$$T_A(\nu) = \frac{\eta}{2 R_L k_B} \int_{\Omega} h_{eff}^2(\nu, \theta, \phi) I_\nu(\nu, \theta, \phi) d\Omega$$

or

$$T_A(\nu) = \frac{\eta \nu^2}{R_L c^2} \int_{\Omega} h_{eff}^2(\nu, \theta, \phi) T_b(\nu, \theta, \phi) d\Omega$$

This is a just a guess to be honest - I would really only use the equation that includes the effective area of the antenna - that's the one in all the textbooks.


### System Gain

$$G_{sys}(\nu) = \frac{P_{cal1}(\nu) - P_{cal2}(\nu)}{T_{cal1}(\nu) - T_{cal2}(\nu)}$$

G<sub>sys</sub>: system gain - dimensionless

P<sub>cal</sub>: spectrometer measurement of the corresponding calibration state - any units as long as its consistant

T<sub>cal</sub>: Calibrator Temperature - K

For High-Z, cal1 is the blackbody (the lightbulb) and cal2 is the 50 $\Omega$ resistor. For both, temperature is constant across all relevant frequencies.

$$G_{sys}(\nu) = \frac{P_{blackbody}(\nu) - P_{resistor}(\nu)}{T_{blackbody} - T_{resistor}}$$

### General Calibration Equation

$$T(\nu) = T_{cal1}(\nu) + \frac{P(\nu) - P_{cal1}(\nu)}{G_{sys}(\nu)}$$

For High-Z

$$T_A(\nu) = T_{blackbody} + \frac{P_A(\nu) - P_{blackbody}(\nu)}{G_{sys}(\nu)}$$

T<sub>A</sub>: Antenna Temperature - K

P<sub>A</sub>: Spectrometer Measurement of the Antenna State - any units as long as its consistant

### System Noise Temperature

The system creates some noise, which we account for in the following way:

$$T_{sys}(\nu) = T_{blackbody} - \frac{P_{blackbody}(\nu)}{G_{sys}(\nu)}$$

### Specific Calibration Equation

$$T_A(\nu) = \frac{P_A(\nu)}{G_{sys}(\nu)} - T_{sys}$$

```python
# compile 2D arrays (spectra for a given time) for each switch state (cleaned/refactored)

state_map = {
    "state0": "antenna",
    "state1": "open_circuit",
    "state2": "short_circuit",
    "state4": "blackbody",
    "state5": "resistor",
    "state6": "noise_diode",
}

# containers for (timestamp, spectrum) pairs
containers = {name: [] for name in state_map.values()}

for file in sorted(data_folder.rglob("*/*.npy")):
    if file.name.startswith("._"):
        continue
    data = np.load(file, allow_pickle=True).item()
    # determine which state this file corresponds to once per file
    state_name = next((name for key, name in state_map.items() if key in file.name), None)
    if state_name is None:
        continue

    for utc_timestamp, state_and_spectrum in data.items():
        try:
            utc_datetime_timestamp = datetime.datetime.strptime(f"{data_folder.name} {utc_timestamp}", "%Y%m%d %H%M%S")
        except ValueError:
            continue
        containers[state_name].append([utc_datetime_timestamp, state_and_spectrum["spectrum"]])

def sort_split_stack(pairs):
    """Sort pairs by timestamp, split into separate lists, and stack spectra into a 2D array."""
    pairs_sorted = sorted(pairs, key=lambda x: x[0])
    timestamps = [p[0] for p in pairs_sorted]
    spectra_list = [p[1] for p in pairs_sorted]
    if spectra_list:
        stacked = np.vstack(spectra_list)
    else:
        stacked = np.empty((0, num_frequency_samples))
    return timestamps, spectra_list, stacked

antenna_utc_timestamps, antenna_spectrum, total_spectrometer_antenna_power = sort_split_stack(containers["antenna"])
blackbody_utc_timestamps, blackbody_spectrum, total_spectrometer_blackbody_power = sort_split_stack(containers["blackbody"])
resistor_utc_timestamps, resistor_spectrum, total_spectrometer_resistor_power = sort_split_stack(containers["resistor"])
noise_diode_utc_timestamps, noise_diode_spectrum, total_spectrometer_noise_diode_power = sort_split_stack(containers["noise_diode"])
open_circuit_utc_timestamps, open_circuit_spectrum, total_spectrometer_open_circuit_power = sort_split_stack(containers["open_circuit"])
short_circuit_utc_timestamps, short_circuit_spectrum, total_spectrometer_short_circuit_power = sort_split_stack(containers["short_circuit"])
```

```python
def rfi_removal(data, threshold=3, median_filter_window=40, passes=3):

    for i in range(passes):
        """ Removes RFI data from the antenna data """
        m = ndimage.median_filter(data, [median_filter_window, median_filter_window])
        logdata = 10*np.log10(data)
        medians = np.median(logdata, axis=0)
        flattened = logdata - medians
        filtered = ndimage.median_filter(flattened, [median_filter_window, median_filter_window])
        corrected = flattened - filtered
        MAD = np.median(np.abs(corrected - np.median(corrected)))
        x = (corrected - np.median(corrected) > threshold*MAD)
        rfi_masked_data = np.ma.masked_array(data, x)
        np.ma.set_fill_value(rfi_masked_data, m)
        data = rfi_masked_data.fill_value

    return data

antenna_rfi_removed = spectrometer_antenna_power#rfi_removal(spectrometer_antenna_power, passes=3)
```

### Load the beam pattern (effective heights)

```python

import numpy as np
import healpy as hp
from scipy.interpolate import RegularGridInterpolator


gain_info = np.loadtxt("mimi_c49_antenna_gain.csv", delimiter=",", skiprows=2)
gain_freqs = np.arange(25, 250+25, 25)
gain_thetas = np.deg2rad(gain_info[:,0])
gain_phis = np.linspace(0, 2*np.pi, 181, endpoint=True)
gain_pattern = gain_info[:,1:].T # each row is a frequency, each column is a theta (elevation)


def load_effective_height(frequency_mhz):
    return gain_pattern[np.where(gain_freqs == frequency_mhz)[0]].repeat(181, axis=0).T

def generate_beam_map(frequency_mhz):
    effective_height_2d_map = load_effective_height(frequency_mhz)
    num_theta, num_phi = effective_height_2d_map.shape
    theta1 = np.linspace(0, np.pi/2, num_theta)
    phi1 = np.linspace(0, 2*np.pi, num_phi)
    interpolator = RegularGridInterpolator(
        (theta1, phi1), effective_height_2d_map,
        bounds_error=False, fill_value=0
    )
    beam_map = np.full(n_pix, 0.0)  # initialize with UNSEEN for masking
    # Get pixel centers in (, )
    theta_hp, phi_hp = hp.pix2ang(n_side, np.arange(n_pix))
    # Find pixels in upper hemisphere ( <= /2)
    mask_upper = theta_hp <= (np.pi / 2)
    # Prepare interpolation points only for those pixels
    interp_points = np.vstack((theta_hp[mask_upper], phi_hp[mask_upper])).T
    # Interpolate
    interp_values = interpolator(interp_points)
    # Assign to map (fill rest with UNSEEN or 0)
    beam_map[mask_upper] = interp_values
    rot = hp.Rotator([0, 90, 0])
    beam_map = rot.rotate_map_pixel(beam_map)
    return beam_map

effective_height_2d_map = load_effective_height(100)

# Create polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
# Set 0 at top, clockwise
ax.set_theta_zero_location('N')  # 0 degrees = North (top)
ax.set_theta_direction(-1)       # Clockwise direction
ax.plot(gain_thetas, effective_height_2d_map[:,0])

plt.title("Basic Polar Plot")
plt.show()

interpolator = RegularGridInterpolator(
    (gain_thetas, gain_phis), effective_height_2d_map,
    bounds_error=False, fill_value=0
)

beam_map = np.full(n_pix, 0.0)  # initialize with UNSEEN for masking

# Get pixel centers in (, )
theta_hp, phi_hp = hp.pix2ang(n_side, np.arange(n_pix))

# Find pixels in upper hemisphere ( <= /2)
mask_upper = theta_hp <= (np.pi / 2)

# Prepare interpolation points only for those pixels
interp_points = np.vstack((theta_hp[mask_upper], phi_hp[mask_upper])).T

# Interpolate
interp_values = interpolator(interp_points)

# Assign to map (fill rest with UNSEEN or 0)
beam_map[mask_upper] = interp_values

rot = hp.Rotator([0, 90, 0])
#theta, phi = rot(theta, phi)
beam_map = rot.rotate_map_pixel(beam_map)
hp.orthview(beam_map, half_sky=True, coord='C', title="Beam Map")
# Now add custom labels for azimuth and zenith angle
ax = plt.gca()

# Add concentric zenith angle circles (like elevation rings)
zenith_angles = [10, 30, 60, 80]
for za in zenith_angles:
    circle = plt.Circle((0, 0), np.sin(np.radians(za)), color='white', ls='--', fill=False, alpha=0.5)
    ax.add_artist(circle)
    plt.text(0, np.sin(np.radians(za)) + 0.01, f"{za}", color='white', ha='center')

# Add azimuth angle labels
az_labels = [0, 90, 180, 270]
label_pos = {
    0: (0, 1.05),       # North (up)
    90: (1.05, 0),      # East (right)
    180: (0, -1.1),     # South (down)
    270: (-1.1, 0),     # West (left)
}
for az in az_labels:
    x, y = label_pos[az]
    plt.text(x, y, f"{az}", color='white', ha='center', va='center')

plt.show()

```
